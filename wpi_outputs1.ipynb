```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel("whole sale price xlsx.xlsx") 


df.dropna(how='all', inplace=True)


df['COMM_WT'].fillna(0, inplace=True)


```

```python

mean_value = df['COMM_WT'].mean()
df['COMM_WT'].fillna(mean_value, inplace=True)

```

```python


# Safely fill NaN values in COMM_WT with the column mean
df.loc[:, 'COMM_WT'] = df['COMM_WT'].fillna(mean_value)


```

```python

mean_value = df['COMM_WT'].mean()
df.loc[:, 'COMM_WT'] = df['COMM_WT'].fillna(mean_value)

```

```python

# Calculate Growth (%) from April 2012 to October 2023

# Calculate Growth (%) from April 2012 to October 2023
df['Growth (%)'] = ((df['INDX102023'] - df['INDX042012']) / df['INDX042012']) * 100

# Get top 10 commodities by growth
top_growth = df[['COMM_NAME', 'Growth (%)']].sort_values(by='Growth (%)', ascending=False).head(10)
print(top_growth)



```

```python
# Plot top 10 commodities by growth
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(top_growth['COMM_NAME'], top_growth['Growth (%)'], color='skyblue')
plt.xlabel('Growth (%)')
plt.title('Top 10 Commodities by WPI Growth (2012–2023)')
plt.gca().invert_yaxis()
plt.grid(True)
plt.tight_layout()
plt.show()
```
